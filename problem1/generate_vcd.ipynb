```python
import os;
```

```python
# call the iverilog compiler
# example : iverilog -o mux_wave multiplexer_2_1.v multiplexer_2_1_tb.v
# go through the train folder and compile all the files

def compile( path ):
    files = os.listdir( path )
    #files = [ f for f in files if f.endswith(".v") ]
    tb_files = [ f for f in files if f.endswith("_tb.v") ]
    progs = [ f for f in files if f.endswith(".v") and f not in tb_files ]

    for i in range( len(progs) ):
        # get the name of the program from tb files
        prog_name = str(i+1) + ".v"
        tb_name = str(i+1) + "_tb.v"

        if prog_name in progs and tb_name in tb_files:
            cmd = "iverilog -o " + path + "/" + str(i+1) + " " +path + "/" + prog_name + " " + path + "/" + tb_name;
            print( cmd );
            os.system( cmd );
```

```python
compile( "train" )
```

```python
# list out the files in train folder that are not .v or .png
def list_files( path ):
    files = os.listdir( path )
    executable_files = [ f for f in files if not f.endswith(".v") and not f.endswith(".png") ]
    print( executable_files )
    return executable_files;
```

```python
executable_files = list_files( "train" )
```

```python
def get_vcd( path, list_of_executables ):
    for f in list_of_executables:
        cmd = "vvp" + " " + path + "/" + f;
        os.system( cmd )
```

```python
get_vcd( "train", executable_files )
```
